```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

# Simulating BankSim transaction data
data = pd.DataFrame({
    "Amount": np.concatenate([np.random.normal(200, 50, 7200), np.random.normal(50, 30, 587443)]),
})

# Model Performance Data
models = ["Logistic Regression", "Random Forest", "Gradient Boosting", "SVM"]
accuracy = [0.9997, 1.0, 1.0, 0.9999]
precision = [0.9824, 1.0, 1.0, 0.9830]
recall = [0.9655, 1.0, 1.0, 1.0]

model_performance = pd.DataFrame({
    "Model": models,
    "Accuracy": accuracy,
    "Precision": precision,
    "Recall": recall
})

# Set style
sns.set_style("whitegrid")

# Create a figure with two subplots (excluding Fraud vs. Non-Fraud chart)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# 1️⃣ Transaction Amount Distribution
sns.histplot(data["Amount"], bins=50, kde=True, ax=axes[0], color="blue")
axes[0].set_title("Distribution of Transaction Amounts")
axes[0].set_xlabel("Transaction Amount ($)")

# 2️⃣ Model Performance Comparison
model_performance.set_index("Model").plot(kind="bar", ax=axes[1], colormap="viridis")
axes[1].set_title("Model Performance Metrics")
axes[1].set_ylabel("Score")

# Adjust layout
plt.tight_layout()

# Save the figure as an image
plt.savefig("bank_reconciliation_updated_charts.png")
plt.show()
```
