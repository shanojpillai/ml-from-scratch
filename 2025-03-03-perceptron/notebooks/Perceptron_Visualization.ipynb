```python
import pandas as pd
import matplotlib.pyplot as plt

# Load experiment results
df_results = pd.read_csv("../data/perceptron_experiment_results.csv")

# Plot accuracy trends
plt.figure(figsize=(8,6))
for eta in df_results["Learning Rate"].unique():
    df_eta = df_results[df_results["Learning Rate"] == eta]
    plt.plot(df_eta["Epochs"], df_eta["Accuracy"], marker="o", linestyle="-", label=f"η={eta}")

plt.xlabel("Epochs")
plt.ylabel("Accuracy (%)")
plt.title("Impact of Learning Rate & Epochs on Perceptron Accuracy")
plt.legend()
plt.grid()
plt.show()

```
